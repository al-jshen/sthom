## Goal: GP + ODE model?

### Some kind of hierarchical model with a MVN prior informed by distances. Hierarchical parameters generate unit level parameters which feed into ODEs. So basically ODE for each unit, but include correlations for parameters of ODEs. 

```python
# generate some cities with (X, Y) locations, populations, initial conditions
n = 5
cities = np.random.normal(size=(n, 2))
population_init = np.random.randint(1e5, 1e6, size=n)
infected_init = np.random.binomial(population_init * (np.linalg.norm(cities, axis=1).sum() / np.linalg.norm(cities, axis=1)).astype(int), 0.00001)
susceptible_init = population_init - infected_init
recovered_init = np.zeros(n)
ic = np.stack((susceptible_init, infected_init, recovered_init))
ic_norm = ic / ic.sum(axis=0)
ic = np.vstack((ic, population_init))
ic_norm = np.vstack((ic_norm, population_init))

plt.scatter(cities[:,0], cities[:,1], s=population_init / 1000, c=np.log(infected_init))
for i in range(n):
    plt.annotate(f"{infected_init[i]} / {population_init[i]}", (cities[i,0], cities[i,1] + 50), fontsize=15)
plt.colorbar()
```

```python
distances = distance_matrix(cities, cities)
plt.matshow(distances)
```

# Modified SIR model v1:

\begin{align}
    \frac{dS_i}{dt} &= -\beta_i S_i I_i T_{i,i} - \sum_{j \neq i} \beta_i S_i I_j T_{ji} + \sum_{j \neq i} S_j T_{ji} - \sum_{j \neq i} S_i T_{ij} \\ 
    \frac{dR_i}{dt} &= \gamma_i I_i T_{i,i} + \sum_{j \neq i} \gamma_i I_j T_{ji} + \sum_{j \neq i} R_j T_{ji} - \sum_{j \neq i} R_i T_{ij} \\
    \frac{dI_i}{dt} &= -\frac{dS_i}{dt} - \frac{dR_i}{dt}
\end{align}

- $\beta_i, \gamma_i$ are the transmission rate and recovery rate in $i$
- $S_i, R_i, I_i$ are the number of susceptible, recovered, and infected in $i$
- $T_{ij}$ is the percentage of the population in $i$ that travels to $j$

# Modified SIR model v2 (simpler):

\begin{align}
    \frac{dS_i}{dt} &= -\beta_i S_i I_i + \sum_{j \neq i} S_j T_{ji} - \sum_{j \neq i} S_i T_{ij} \\ 
    \frac{dR_i}{dt} &= \gamma_i I_i + \sum_{j \neq i} R_j T_{ji} - \sum_{j \neq i} R_i T_{ij} \\
    \frac{dI_i}{dt} &= -\frac{dS_i}{dt} - \frac{dR_i}{dt}
\end{align}

```python
tgt_travel = 0.05

travels = distances / distances.sum()
for i in range(n):
    travels[i,i] = 1.
travels = 1. / travels
travels /= travels.sum()
travels += np.random.normal(0, 0.1, size=travels.shape)**2
travels /= np.median(travels[travels != 1]) / tgt_travel

for i in range(n):
    travels[i,i] = 1 - tgt_travel
    mask = np.ones(n).astype(bool)
    mask[i] = False
    travels[i][mask] = travels[i][mask] / travels[i][mask].sum() * tgt_travel

travels[np.isnan(travels)] = 0.
    
plt.matshow(np.clip(travels, None, 0.05))
plt.colorbar()
```

```python
betas = np.random.uniform(0.1, 2, size=n)
gammas = np.random.triangular(0, 0, 0.5, size=n)**2


fig = plt.figure(figsize=(20, 6))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)
ax1.hist(betas)
ax2.hist(gammas)
ax3.hist(travels[travels < 0.5].flatten())
```

```python
days = 100
resolution = 5
timesteps = days * resolution

y = np.zeros((timesteps + 1, *ic.shape))
y[0] = ic_norm

for d in range(timesteps):
    
    dydt = np.zeros_like(ic)
        
    for i in range(n):
        mask = np.ones(n).astype(bool)
        mask[i] = False
        
        # v1
        dydt[0,i] = - betas[i] * y[d,0,i] * y[d,1,i] * travels[i,i] - np.sum(betas[i] * y[d,0][mask] * y[d,1][mask] * travels[i][mask]) + np.sum(y[d,0][mask] * travels[i][mask]) - np.sum(y[d,0,i] * travels[i][mask])
        dydt[2,i] = gammas[i] * y[d,1,i] * travels[i,i] + np.sum(gammas[i] * y[d,1][mask] * travels[i][mask]) + np.sum(y[d,2][mask] * travels[i][mask]) - np.sum(y[d,2,i] * travels[i][mask])
        
        # v2
        # dydt[0,i] = - betas[i] * y[d,0,i] * y[d,1,i] + np.sum(y[d,0][mask] * y[d,1][mask] * travels[i][mask]) - np.sum(y[d,0,i] * travels[i][mask])
        # dydt[2,i] = gammas[i] * y[d,1,i] + np.sum(y[d,2][mask] * travels[i][mask]) - np.sum(y[d,2,i] * travels[i][mask])
        
        # reference
        # dydt[0,i] = - betas[i] * y[d,0,i] * y[d,1,i]
        # dydt[2,i] = gammas[i] * y[d,1,i]
        
        dydt[1,i] = -dydt[0,i] - dydt[2,i]
                
    y[d + 1] = y[d] + dydt / resolution
```

```python
t = np.linspace(0, days, timesteps + 1)
for i in range(n):
    plt.plot(t, y[:,0,i], c='y')
    plt.plot(t, y[:,1,i], c='r')
    plt.plot(t, y[:,2,i], c='g')
```

```python
sample_days = 25 # number of data points
sample_n = 50 # number of people sampled per day

obs_sample_times = np.zeros((n, sample_days))
obs_infected_count = np.zeros((n, sample_days))

for i in range(n):

    sample_times = np.zeros(sample_days)
    sample_times = np.random.choice(t[1:] * resolution, size=sample_days, replace=False) # only sampled during these days
    sample_times = np.sort(sample_times)
    sample_true_i = y[:,1,i][sample_times.astype(int)] # true % infected on sampled days
    sample_i = np.random.binomial(sample_n, sample_true_i, size=sample_days) # binomial error on measurements

    obs_sample_times[i] = sample_times
    obs_infected_count[i] = sample_i
    
    # plt.scatter(sample_times, sample_true_i)
    plt.scatter(sample_times, sample_i / sample_n, label='Simulated measurements')
    plt.plot(t * resolution, y[:,1,i], label='True infections')
    # plt.legend(fontsize=20)
    plt.xlabel('t')
    plt.ylabel('% of population')
```

```python
sm = cmdstanpy.CmdStanModel(stan_file='./model.stan')
```

```python
standata = dict(
    n = n,
    n_obs = sample_days,
    n_sample = sample_n,
    distances = distances,
    t_obs = obs_sample_times,
    infected_obs = obs_infected_count.astype(int)
)
```

```python
fit = sm.sample(standata, chains=4, parallel_chains=4, show_progress='notebook')
```
